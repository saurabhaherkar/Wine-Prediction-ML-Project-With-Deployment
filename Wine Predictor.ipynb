```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
x = df.drop(['target'], axis=1)
y = df['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=51)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Shortcut method for finding the accuracy 

```python
model = []
model.append(('LR', LogisticRegression()))
model.append(('LDA', LinearDiscriminantAnalysis()))
model.append(('KNN', KNeighborsClassifier()))
model.append(('CART', DecisionTreeClassifier()))
model.append(('NB', GaussianNB()))
model.append(('SVM', SVC()))
print(model)
```

```python
results = []
names = []
for name, model in model:
    kfold = KFold(n_splits=10, random_state=51)
    cv_result = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')
    results.append(cv_result)
    names.append(name)
    msg = '{}: {}, ({})'.format(name, cv_result.mean(), cv_result.std())
    print(msg)
```

# Longcut method for predicting the score and finding the accuracy

```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(x_train, y_train)
prediction_lr = lr.predict(x_test)
print(accuracy_score(y_test, prediction_lr))
print(classification_report(y_test, prediction_lr))
print(confusion_matrix(y_test, prediction_lr))
```

```python
# Decision Tree Classification
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
prediction_dt = dt.predict(x_test)
print(accuracy_score(y_test, prediction_dt))
print(classification_report(y_test, prediction_dt))
print(confusion_matrix(y_test, prediction_dt))
```

```python
# K Nearest Neighbors Classification
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
prediction_knn = knn.predict(x_test)
print(accuracy_score(y_test, prediction_knn))
print(classification_report(y_test, prediction_knn))
print(confusion_matrix(y_test, prediction_knn))
```

```python
# GuassianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
prediction_nb = nb.predict(x_test)
print(accuracy_score(y_test, prediction_nb))
print(classification_report(y_test, prediction_nb))
print(confusion_matrix(y_test, prediction_nb))
```

```python
# Support Vector Classification
svc = SVC()
svc.fit(x_train, y_train)
prediction_svc = svc.predict(x_test)
print(accuracy_score(y_test, prediction_svc))
print(classification_report(y_test, prediction_svc))
print(confusion_matrix(y_test, prediction_svc))
```

```python
# Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
prediction_lda = lda.predict(x_test)
print(accuracy_score(y_test, prediction_lda))
print(classification_report(y_test, prediction_lda))
print(confusion_matrix(y_test, prediction_lda))
prediction_lda
```
